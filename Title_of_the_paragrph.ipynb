```python
import urllib.request
```

```python
from bs4 import BeautifulSoup
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
response=urllib.request.urlopen('https://en.wikipedia.org/wiki/MS_Dhoni')
html=response.read()
```

```python
print(html)
```

```python
soup=BeautifulSoup(html,'html5lib')
```

```python
print(soup)
```

```python
text=soup.get_text(strip=True)
```

```python
print(text)
```

```python
tokens=[t for t in text.split()]
print(tokens)
```

```python
sr=stopwords.words('english')
print(sr)
```

```python
clean_words=tokens[:]
for token in tokens[:]:
    if token in sr:
        clean_words.remove(token)
print(clean_words)        
```

```python
freq=nltk.FreqDist(clean_words)
print(freq)
```

```python
for key,val in freq.items():
    print(str(key)+':'+str(val))
```

```python
freq.plot(20,cumulative=False)
```

```python
key_list = list(freq.keys()) 
val_list = list(freq.values()) 
```

```python
print(max(val_list))
```

```python
s=max(val_list)
```

```python
print("The title of the above wikipedia is"+ " "+key_list[val_list.index(s)])
```
